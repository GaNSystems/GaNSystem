```python
# Function of output capacitance charge/discharge
def Ecoss(Coss,Vds):
    return 0.5 * (2 * Coss) * Vds ** 2
# Reverse conduction loss calculation
def Ercon(Vgs_th,Vgs_off,Rdson,Id,t_d):
    return (-Vgs_off + Vgs_th + Rdson ** 2) * Id * t_d
Vout = 400
Vin = 300
fsw = 100e3
T = 1 / fsw
Vgs_off = 0
Vgs_th = 1.3
D = 1 - Vin / Vout
L = 1e-5
P = 8000
Iout = P / Vout
I_inductor = Iout / (1 - D)
I_inductor_pp = (Vout - Vin) * (1 - D) / (L * fsw)
dI_inductor_down = (Vout - Vin) / L
dI_inductor_up = Vin / L
Rdson = 50e-3
Coss = 65e-12
Coss_energyrelated = 88e-12
Ecoss = Ecoss(Coss,Vout)


t_d = Coss_energyrelated * Vout / (I_inductor * 0.5)
Ercon = Ercon(Vgs_th,Vgs_off,Rdson,I_inductor,t_d)

from pandas import Series, DataFrame
Eloss ={'Ecoss(uJ)':[Ecoss*1e6],'Ercon(uJ)':[Ercon*1e6],'Eonnew_Id25(uJ)':[Eonnew_Id25]}
Eloss_table = DataFrame(Eloss) # creating a data frame
Eloss_table
```
